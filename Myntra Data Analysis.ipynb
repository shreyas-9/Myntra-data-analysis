# 3. Data Analysis - Numeric Column

#### Average Rating of a product (non zero)

```python
rat = 0
c = 0
for i in df['rating']:
    if(i != 0.0):
        rat += i
        c += 1
        
print(round(int(rat) / c,2))
```

## 3.8) Finding the top 10 brands with most number of products

```python
data = []


for brand_name in df['brand_name'].unique():
    c = 0
    for i in df['brand_name']:
        if (i == brand_name):
            c += 1

    data.append([brand_name , c])
    
    
data = pd.DataFrame(data, columns = ['name','products'])

data.sort_values('products', ascending = False).head(10)
```

```python
import numpy as np

pincode = []
state = []

for i in df['seller_address']:
    
    if (i != 'Not Mentioned'):
        pincode.append(i.split(' ')[-1])
        state.append(i.split(',')[-1].split('-')[0].strip())                 #strip() to remove extra spaces
    else:
        pincode.append(np.nan)
        state.append(np.nan)
        
df['pincode'] = pincode
df['state'] = state
```
